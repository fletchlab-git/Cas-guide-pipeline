```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Feb 11 14:41:39 2025

@author: zainamoussa
"""

from Bio import Entrez
import pandas as pd
import re

# Input file containing the BLAST results (assumed to be tab-delimited or space-separated)
input_file = "filtered_BLAST_topguides_gen_crossreactivity_donovani_results.txt"
# Read the file into a pandas DataFrame
blast_results = pd.read_csv(input_file, delim_whitespace=True)

# Add new columns for TaxID and Scientific Name
blast_results['Subject_TaxID'] = None  
blast_results['Scientific_Name'] = None  

# Set your email for NCBI Entrez (required)
Entrez.email = "zaina_moussa@berkeley.edu"

# Loop through each accession in the 'Offtarget' column
p = 0
for accession in blast_results['Offtarget']:
    accession = accession.split()[0]  # Extract just the accession part
    
    try:
        # Fetch the summary which includes TaxID and Title (contains the scientific name)
        handle = Entrez.esummary(db="nucleotide", id=accession, retmode="xml")
        records = Entrez.read(handle)
        
        # Extract TaxID and Title (contains scientific name)
        taxid = records[0].get("TaxId", "Not found")
        title = records[0].get("Title", "Not found")

        # Store the TaxID

        blast_results.at[p, 'Subject_TaxID'] = re.search(r"IntegerElement\((\d+)", str(taxid)).group(1)
        
        # Extract the scientific name from the title if TaxID was found
        if taxid != "Not found" and title != "Not found":
            sci_name = title.split('[')[-1].rstrip(']')
            blast_results.at[p, 'Scientific_Name'] = sci_name
        
    except Exception as e:
        print(f"Error fetching data for {accession}: {e}")
    
    print(f"Processed {p+1}/{len(blast_results)}")
    p += 1  # Increment the counter

# Save the updated DataFrame to a new file
blast_results.to_csv("blast_cross_reactivity_donovani_results_with_taxids.txt", sep="\t", index=False)
```
